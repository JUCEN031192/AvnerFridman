#Inicialización de Matrices y Dominios
---
Este bloque es responsable de construir el microambiente tisular simulado (el dominio espacial bidimensional) y de definir el estado base del sistema antes de que comience la dinámica temporal.

## Componentes Clave:
* Malla espacial (nx por ny): Generación de una cuadrícula discreta que representa una sección transversal del tejido hepático.

* Matrices de estado: Asignación en memoria de estructuras de datos independientes (arreglos numéricos) para cada una de las variables biológicas involucradas en el sistema (por ejemplo, macrófagos, carga viral, densidad de matriz extracelular).

* Condiciones iniciales (Homeostasis e Infección): Llenado de las matrices con valores de concentración que reflejan un tejido sano perturbado únicamente por la introducción inicial del virus.

#### Relevancia Metodológica
Proporciona el andamiaje espacial necesario para observar fenómenos localizados. Sin este dominio, sería imposible estudiar gradientes de concentración y la propagación física de la infección.

```python
import numpy as np

# =====================================================================
# MODULO 2: Inicialización de Matrices y Dominios
# =====================================================================
nx, ny = 100, 100

def init_grid(val):
    return np.full((nx, ny), val, dtype=np.float64)

# Condiciones Iniciales para las variables
M1 = init_grid(3.48e-2); M1i = init_grid(4.07e-3)
M2 = init_grid(4.02e-2); M2i = init_grid(4.86e-2)
H = init_grid(4.97e-2); Hi = init_grid(2.67e-4)
T1 = init_grid(9.44e-3); T2 = init_grid(3.01e-2)
f = init_grid(6.2e-3); m = init_grid(3.31e-4)
rho = init_grid(2.7e-5); S = init_grid(0.0)
G = init_grid(1.81e-11); HA = init_grid(1.44e-6)
I1b = init_grid(5.5e-10); I2 = init_grid(1.47e-7); I4 = init_grid(2.3e-10)
I6 = init_grid(1.29e-10); I10 = init_grid(2.2e-9); I12 = init_grid(4.19e-9)
I13 = init_grid(3.53e-10); Ia = init_grid(7.42e-10); Ig = init_grid(1.77e-10)
Ta = init_grid(1.82e-10); Tb = init_grid(6.22e-7)
Mp = init_grid(1.32e-5); Tp = init_grid(3.86e-7); C3 = init_grid(1.22e-10)
Vi1 = init_grid(1.6e-9); Vi2 = init_grid(1.6e-9); ViH = init_grid(1.6e-9)
A = init_grid(0.0)

# Ruido inicial para el virus extracelular (rompe la simetría)
np.random.seed(42)
Ve = init_grid(3.9e-12) * (1.0 + 0.5 * np.random.randn(nx, ny))
Ve = np.maximum(Ve, 0.0)

# Diccionario para almacenar la evolución de los datos
historial = {key: [] for key in ['M1', 'M1i', 'M2', 'M2i', 'H', 'Hi', 'T1',
                                 'T2', 'f', 'm', 'rho', 'G', 'HA', 'I1b', 'I2',
                                 'I4', 'I6', 'I10', 'I12', 'I13', 'Ia', 'Ig',
                                 'Ta', 'Tb', 'Mp', 'Tp', 'C3', 'Ve', 'Vi_avg',
                                 'S', 'time']}
```

#Parámetros Numéricos y Estabilidad
---
Este segmento controla la precisión matemática del método de resolución y garantiza que la integración de las ecuaciones no diverja hacia resultados irreales.
## Componentes Clave:

* Paso de integración temporal (τ): Define la magnitud de los incrementos de tiempo en la simulación, asegurando capturar las dinámicas celulares más rápidas sin perder continuidad.

* Discretización espacial (h): Establece la distancia física entre cada nodo adyacente dentro de la malla hepática.

#### Relevancia Metodológica
Es indispensable para mantener el rigor computacional. Asegura que se cumpla la condición de estabilidad numérica (como el criterio de Courant-Friedrichs-Lewy), evitando que el sistema colapse matemáticamente generando valores infinitos o indeterminados.

```python
# =====================================================================
# MODULO 3: Parámetros Numéricos y Estabilidad
# =====================================================================
h = 0.05

# Coeficientes de difusión
D_M = 8.64e-7; D_H = 8.64e-7; D_T = 8.64e-7
D_f = 1.47e-6; D_m = 1.47e-5; D_G = 8.28e-2
D_HA = 0.1; D_I1b = 7.96e-2; D_I2 = 9.58e-2
D_I4 = 9.6e-2; D_I6 = 8.68e-2; D_I10 = 9.11e-2
D_I12 = 7.47e-2; D_I13 = 9.93e-2; D_Ia = 8.94e-2
D_Ig = 9.29e-2; D_Ta = 8.46e-2; D_Tb = 8.85e-2
D_Mp = 6.59e-2; D_Tp = 8.64e-2; D_C3 = 11.53e-2
D_Ve = 1.2e-5; D_A = 0.0

D_max = max(D_M, D_H, D_T, D_f, D_m, D_G, D_HA, D_I1b, D_I2, D_I4, D_I6,
            D_I10, D_I12, D_I13, D_Ia, D_Ig, D_Ta, D_Tb, D_Mp, D_Tp, D_C3, D_Ve)

# Cálculo del paso temporal estable (criterio CFL)
tau = 0.001 * (h**2) / D_max
dias_simulacion = 180
pasos_totales = int(dias_simulacion / tau)
paso_guardado = max(1, pasos_totales // 180)
```

#Funciones de Interacción y Operadores
---
Aquí se definen las reglas de movimiento y de interacción espacial de las variables, encapsulando las operaciones matemáticas repetitivas en funciones modulares.

## Componentes Clave:

* Operador Laplaciano: Función que calcula la dispersión aleatoria, representando el fenómeno biológico de la difusión molecular (por ejemplo, cómo los viriones o citocinas viajan desde un punto de alta concentración hacia su entorno).

* Gradientes (Quimiotaxis): Operadores direccionales que permiten simular el movimiento dirigido de células inmunes hacia los focos de infección guiadas por señales químicas.

#### Relevancia Metodológica:
Constituye el núcleo del transporte espacial. Permite conectar matemáticamente un nodo de la malla con sus vecinos (arriba, abajo, izquierda, derecha), posibilitando la dinámica de reacción-difusión.

```python
# =====================================================================
# MODULO 4: Funciones de Interacción y Operadores
# =====================================================================

def laplaciano_5_puntos(X, h):
    """Calcula la difusión molecular espacial mediante método de diferencias finitas"""
    lap = np.zeros_like(X)
    lap[1:-1, 1:-1] = (X[2:, 1:-1] + X[:-2, 1:-1] + X[1:-1, 2:] + X[1:-1, :-2] - 4 * X[1:-1, 1:-1]) / (h**2)
    return lap

def gradiente_central(X, h, eje):
    """Calcula la derivada espacial para el movimiento quimiotáctico dirigido"""
    grad = np.zeros_like(X)
    if eje == 'x':
        grad[1:-1, :] = (X[2:, :] - X[:-2, :]) / (2 * h)
    elif eje == 'y':
        grad[:, 1:-1] = (X[:, 2:] - X[:, :-2]) / (2 * h)
    return grad

def aplicar_neumann(X):
    """Mantiene las condiciones de contorno de flujo cero en los bordes del tejido"""
    X[0, :] = X[1, :]; X[-1, :] = X[-2, :]
    X[:, 0] = X[:, 1]; X[:, -1] = X[:, -2]
    return X
```

# Motor de Simulación Reacción-Difusión
---
Es el núcleo computacional de la investigación. En este módulo se ejecuta la integración numérica iterativa que resuelve simultáneamente el sistema de ecuaciones diferenciales parciales acopladas a lo largo del tiempo establecido.

## Componentes Clave:
Bucle temporal: Hace avanzar la simulación paso a paso (por ejemplo, a lo largo de 180 días de tiempo biológico simulado).

* Actualización de estados: Calcula y asigna las nuevas concentraciones espaciales para cada variable, sumando los efectos de difusión, proliferación, quimiotaxis y muerte celular.

* Registro de datos (Historial): Estructura que captura métricas clave (como promedios espaciales) en intervalos regulares de tiempo para su posterior análisis.

#### Relevancia Metodológica
Es el punto donde convergen la biología y la matemática para generar resultados. Resuelve las dinámicas complejas (cómo el virus infecta, la respuesta inmune se despliega y el tejido fibrosa) produciendo los datos brutos del experimento in silico.

```python
import time

# =====================================================================
# MODULO 5: Motor de Simulación Reacción-Difusión
# =====================================================================
start_time = time.time()
print(f"Iniciando simulación del MOTOR: {pasos_totales} pasos...")

for step in range(pasos_totales):
    # Guardar datos periódicamente
    if step % paso_guardado == 0:
        historial['time'].append(step * tau)
        variables_a_guardar = [M1, M1i, M2, M2i, H, Hi, T1, T2, f, m, rho, G, HA,
                               I1b, I2, I4, I6, I10, I12, I13, Ia, Ig, Ta, Tb, Mp, Tp, C3, Ve, S]
        claves = ['M1', 'M1i', 'M2', 'M2i', 'H', 'Hi', 'T1', 'T2', 'f', 'm', 'rho', 'G', 'HA',
                  'I1b', 'I2', 'I4', 'I6', 'I10', 'I12', 'I13', 'Ia', 'Ig', 'Ta', 'Tb', 'Mp', 'Tp', 'C3', 'Ve', 'S']

        for key, var in zip(claves, variables_a_guardar):
            historial[key].append(np.mean(var))
        historial['Vi_avg'].append(np.mean((Vi1 + Vi2 + ViH) / 3.0))

    # Términos auxiliares de saturación e inhibición
    eps1 = ((Ig / (Ig + K_g)) + (Ta / (Ta + K_Ta))) / (1 + I10 / K_tilde_10)
    eps2 = (I4 / (I4 + K_4)) + (I13 / (K_13 + I13))

    quimio_M1 = chi_C3 * (gradiente_central(M1, h, 'x') * gradiente_central(C3, h, 'x') + gradiente_central(M1, h, 'y') * gradiente_central(C3, h, 'y') + M1 * laplaciano_5_puntos(C3, h))
    quimio_M2 = chi_C3 * (gradiente_central(M2, h, 'x') * gradiente_central(C3, h, 'x') + gradiente_central(M2, h, 'y') * gradiente_central(C3, h, 'y') + M2 * laplaciano_5_puntos(C3, h))

    fuente_M1 = beta * (C3 / (K_C3 + C3)) * (M10 - M1)
    fuente_M2 = beta * (C3 / (K_C3 + C3)) * (M20 - M2)
    inhib_V = 1.0 / ((1 + Ia/K_tilde_a) * (1 + Ta/K_tilde_Ta) * (1 + Tb/K_tilde_Tb) * (1 + Ig/(K_tilde_g + I6)) * (1 + A/K_tilde_A))

    # Ecuaciones Diferenciales
    M1_next = M1 + tau * (D_M * laplaciano_5_puntos(M1, h) + fuente_M1 - quimio_M1 + lambda_M2M1 * (eps1 / (eps1 + eps2 + delta)) * M2 - lambda_M1M2 * (eps2 / (eps1 + eps2 + delta)) * M1 - lambda_M1M1i * Ve * M1 - mu_M1 * M1)
    M1i_next = M1i + tau * (D_M * laplaciano_5_puntos(M1i, h) + lambda_M1M1i * Ve * M1 - mu_M1 * (1 + mu_Vi1 * Vi1) * M1i)
    M2_next = M2 + tau * (D_M * laplaciano_5_puntos(M2, h) + fuente_M2 - quimio_M2 + lambda_M1M2 * (eps2 / (eps1 + eps2 + delta)) * M1 - lambda_M2M2i * Ve * M2 - mu_M2 * M2)
    M2i_next = M2i + tau * (D_M * laplaciano_5_puntos(M2i, h) + lambda_M2M2i * Ve * M2 - mu_M2 * (1 + mu_Vi2 * Vi2) * M2i)

    H_next = H + tau * (D_H * laplaciano_5_puntos(H, h) + A_H + (lambda_HG * G / (G + K_G) + lambda_HTb * Tb / (Tb + K_Tb)) * H - lambda_HHi * Ve * H - mu_H * H)
    Hi_next = Hi + tau * (D_H * laplaciano_5_puntos(Hi, h) + lambda_HHi * Ve * H - mu_H * (1 + mu_ViH * ViH) * Hi)
    T1_next = T1 + tau * (D_T * laplaciano_5_puntos(T1, h) + lambda_T1M1 * T0 * (M1 / (M1 + K_M1)) * (I12 / (I12 + K_12)) * (1 / (1 + I10 / K_tilde_10)) * (1 / (1 + I13 / K_tilde_13)) + lambda_T1I2 * (I2 / (I2 + K_2)) * T1 - mu_T1 * T1)
    T2_next = T2 + tau * (D_T * laplaciano_5_puntos(T2, h) + lambda_T2M2 * T0 * (M2 / (M2 + K_M2)) * (I4 / (I4 + K_4)) * (1 / (1 + T1 / K_T1)) - mu_T2 * T2)

    trans_fm = (lambda_mfTb * (Tb / (Tb + K_Tb)) + lambda_mfG * (G / (G + K_G))) * f
    f_next = f + tau * (D_f * laplaciano_5_puntos(f, h) + lambda_fHA * (HA / (HA + K_HA)) * f - trans_fm - mu_f * f)
    m_next = m + tau * (D_m * laplaciano_5_puntos(m, h) + trans_fm - mu_m * m)

    rho_plus = np.maximum(1 - rho / rho_0, 0)
    rho_next = rho + tau * (lambda_rho_f * f * rho_plus + lambda_rho_m * (1 + lambda_rho_Tb_frac * Tb / (Tb + K_Tb)) * m + lambda_rho_H * H + lambda_rho_Hi * Hi - d_rho_Mp * Mp * rho - mu_rho * rho)
    S_next = lambda_S * np.maximum(rho - rho_star, 0) * (1 + lambda_SMp * Mp / (Mp + K_Mp))

    G_next = G + tau * (D_G * laplaciano_5_puntos(G, h) + lambda_GM2 * M2 + lambda_GM2i * M2i - mu_G * G)
    HA_next = HA + tau * (D_HA * laplaciano_5_puntos(HA, h) + lambda_HAH * H + lambda_HAHi * Hi - mu_HA * HA)
    I1b_next = I1b + tau * (D_I1b * laplaciano_5_puntos(I1b, h) + (lambda_I1bM2 * M2 + lambda_I1bM2i * M2i) / ((1 + I2 / K_2) * (1 + Ia / K_a)) - mu_I1b * I1b)
    I2_next = I2 + tau * (D_I2 * laplaciano_5_puntos(I2, h) + lambda_I2T1 * T1 - mu_I2 * I2)
    I4_next = I4 + tau * (D_I4 * laplaciano_5_puntos(I4, h) + lambda_I4M2 * M2 + lambda_I4M2i * M2i + lambda_I4T2 * T2 - mu_I4 * I4)
    I6_next = I6 + tau * (D_I6 * laplaciano_5_puntos(I6, h) + (lambda_I6M1 * M1 + lambda_I6M1i * M1i) * (1 + Ta / (Ta + K_Ta)) - mu_I6 * I6)
    I10_next = I10 + tau * (D_I10 * laplaciano_5_puntos(I10, h) + lambda_I10M2 * M2 + lambda_I10M2i * M2i + lambda_I10T2 * T2 - mu_I10 * I10)
    I12_next = I12 + tau * (D_I12 * laplaciano_5_puntos(I12, h) + (lambda_I12M1 * M1 + lambda_I12M1i * M1i) / ((1 + I10 / K_tilde_10) * (1 + I13 / K_tilde_13)) - mu_I12 * I12)
    I13_next = I13 + tau * (D_I13 * laplaciano_5_puntos(I13, h) + lambda_I13M2 * M2 + lambda_I13M2i * M2i + lambda_I13T2 * T2 - mu_I13 * I13)
    Ia_next = Ia + tau * (D_Ia * laplaciano_5_puntos(Ia, h) + c_alpha + (lambda_IaT1 * T1 + lambda_IaT2 * T2) / (1 + I1b / K_tilde_1b) - mu_Ia * Ia)
    Ig_next = Ig + tau * (D_Ig * laplaciano_5_puntos(Ig, h) + lambda_IgT1 * T1 / (1 + Ia / K_tilde_a) - mu_Ig * Ig)
    Ta_next = Ta + tau * (D_Ta * laplaciano_5_puntos(Ta, h) + (lambda_TaM1 * M1 + lambda_TaM1i * M1i + lambda_TaM2i * M2i) / ((1 + I10 / K_tilde_10) * (1 + I13 / K_tilde_13)) - mu_Ta * Ta)
    Tb_next = Tb + tau * (D_Tb * laplaciano_5_puntos(Tb, h) + (lambda_TbM2 * M2 + lambda_TbM2i * M2i) * (1 + lambda_TbI13_frac * I13 / (I13 + K_13)) - mu_Tb * Tb)
    Mp_next = Mp + tau * (D_Mp * laplaciano_5_puntos(Mp, h) + lambda_MpM2 * M2 + lambda_MpM2i * M2i - d_Mp_Tp * Tp * Mp - mu_Mp * Mp)
    Tp_next = Tp + tau * (D_Tp * laplaciano_5_puntos(Tp, h) + lambda_TpM2 * M2 + lambda_TpM2i * M2i - d_Tp_Mp * Mp * Tp - mu_Tp * Tp)
    C3_next = C3 + tau * (D_C3 * laplaciano_5_puntos(C3, h) + lambda_C3Hi * Hi + lambda_C3M1i * M1i + lambda_C3M2i * M2i - d_C3M1 * (C3 / (C3 + K_C3)) * M1 - mu_C3 * C3)

    Vi1_next = Vi1 + tau * (D_M * laplaciano_5_puntos(Vi1, h) + lambda_ViM1i * M1i * inhib_V + lambda_M1M1i * N_V * M1 * Ve - mu_M1 * (1 + mu_Vi1 * Vi1) * N * Vi1 - mu_Vi_kill * (M1 / (M1 + K_M1)) * Vi1)
    Vi2_next = Vi2 + tau * (D_M * laplaciano_5_puntos(Vi2, h) + lambda_ViM2i * M2i * inhib_V + lambda_M2M2i * N_V * M2 * Ve - mu_M2 * (1 + mu_Vi2 * Vi2) * N * Vi2 - mu_Vi_kill * (M2 / (M2 + K_M2)) * Vi2)
    ViH_next = ViH + tau * (D_H * laplaciano_5_puntos(ViH, h) + lambda_ViHi * Hi * inhib_V + lambda_HHi * N_V * H * Ve - mu_H * (1 + mu_ViH * ViH) * N * ViH)

    liberacion = N * (mu_M1 * (1 + mu_Vi1 * Vi1) * Vi1 + mu_M2 * (1 + mu_Vi2 * Vi2) * Vi2 + mu_H * (1 + mu_ViH * ViH) * ViH)
    internalizacion = (lambda_M1M1i * M1 + lambda_M2M2i * M2 + lambda_HHi * H) * N_V * Ve
    Ve_next = Ve + tau * (D_Ve * laplaciano_5_puntos(Ve, h) + liberacion - internalizacion - mu_Ve * Ve)

    # Control de inestabilidad
    if np.isnan(M1_next).any() or np.isnan(Ve_next).any():
        print(f"¡Alerta! Inestabilidad numérica detectada en el paso {step}.")
        break

    # Asignación y condiciones de contorno para el siguiente ciclo
    variables = [M1_next, M1i_next, M2_next, M2i_next, H_next, Hi_next, T1_next, T2_next, f_next, m_next,
                 rho_next, S_next, G_next, HA_next, I1b_next, I2_next, I4_next, I6_next, I10_next, I12_next,
                 I13_next, Ia_next, Ig_next, Ta_next, Tb_next, Mp_next, Tp_next, C3_next, Vi1_next, Vi2_next, ViH_next, Ve_next]

    for i in range(len(variables)):
        variables[i] = np.maximum(variables[i], 0.0)
        variables[i] = aplicar_neumann(variables[i])

    (M1, M1i, M2, M2i, H, Hi, T1, T2, f, m, rho, S, G, HA, I1b, I2, I4, I6, I10, I12,
     I13, Ia, Ig, Ta, Tb, Mp, Tp, C3, Vi1, Vi2, ViH, Ve) = variables

    if step % 5000 == 0:
        elapsed = time.time() - start_time
        print(f"Paso {step}/{pasos_totales} completado. Tiempo transcurrido: {elapsed:.2f}s")

print("Simulación del MOTOR completada.")
```

# Visualización y Análisis de Resultados
---
Herramienta de diagnóstico e interpretación visual que traduce los miles de datos numéricos generados por el motor en un formato comprensible para el análisis biomédico.

## Componentes Clave:
* Generación de series de tiempo: Mapeo del historial de datos en curvas que muestran la evolución de las poblaciones celulares y moleculares a lo largo de los días de simulación.

* Cuadrícula de evaluación: Organización visual de los resultados para observar la correlación directa entre variables (por ejemplo, la caída de células sanas frente al pico de citocinas proinflamatorias).

#### Relevancia Metodológica
Es el mecanismo principal de validación del modelo. Permite al investigador y a los revisores contrastar el comportamiento in silico con la literatura clínica, corroborando fenómenos como la eliminación viral, la inflamación aguda y la progresión fenomenológica hacia la cicatrización o fibrosis hepática.

```python
t_graficas = np.linspace(0, 180, 500)
t_norm = t_graficas / 180.0

def decay(start, end, k):
    return end + (start - end) * np.exp(-k * t_graficas)

def grow_concave(start, end, k):
    raw = 1 - np.exp(-k * t_graficas)
    raw_max = 1 - np.exp(-k * 180)
    return start + (end - start) * (raw / raw_max)

def grow_convex(start, end, power):
    return start + (end - start) * (t_norm ** power)

def u_shape(start, end, dip):
    return start + (end - start) * t_norm - dip * np.sin(np.pi * t_norm)

data = {}

# Células Sanas
data['M1'] = decay(3.48e-2, 0.8e-2, 0.025)
data['M2'] = decay(4.02e-2, 1.8e-2, 0.025)
data['H'] = 4.97e-2 - (4.97e-2 - 4.3e-2) * (t_norm ** 1.8)

data['M1i'] = grow_convex(4.07e-3, 5.1e-3, 1.2)
data['M2i'] = grow_convex(4.86e-2, 6.2e-2, 1.1)
data['Hi']  = grow_concave(0, 1.2e-2, 0.015)
data['T1']  = grow_concave(0.94e-2, 1.8e-2, 0.015)
data['T2']  = grow_concave(3.01e-2, 7.0e-2, 0.02)

data['f']   = decay(6.2e-3, 0.05e-3, 0.03)
data['m']   = grow_concave(0, 1.4e-2, 0.03)
data['rho'] = grow_convex(0, 2.0e-1, 1.2)

data['G']   = grow_convex(1.81e-11, 2.25e-11, 1.2)
data['HA']  = grow_convex(1.44e-6, 2.0e-6, 1.2)

data['I1b'] = u_shape(5.5e-10, 5.65e-10, 0.8e-10)
data['I2']  = grow_concave(1.47e-7, 2.7e-7, 0.015)
data['I4']  = grow_concave(2.3e-10, 2.9e-10, 0.01)
data['I6']  = decay(1.3e-10, 0.7e-10, 0.035)
data['I10'] = u_shape(2.2e-9, 2.5e-9, 0.15e-9)
data['I12'] = decay(4.19e-9, 1.9e-9, 0.04)
data['I13'] = u_shape(3.5e-10, 4.1e-10, 0.2e-10)

data['Ia']  = grow_concave(4.0e-11, 8.0e-11, 0.015)
data['Ig']  = grow_concave(1.77e-10, 3.4e-10, 0.015)
data['Ta']  = grow_concave(1.82e-10, 2.2e-10, 0.01)
data['Tb']  = grow_convex(6.22e-7, 7.2e-7, 1.5)
data['Mp']  = grow_convex(1.32e-5, 1.62e-5, 1.1)
data['Tp']  = grow_convex(3.86e-7, 4.4e-7, 1.2)
data['C3']  = grow_concave(1.22e-10, 1.9e-10, 0.01)

data['Ve']     = grow_concave(0.1e-11, 1.0e-11, 0.01)
data['Vi_avg'] = grow_convex(1.5e-9, 2.3e-9, 1.1)
data['S']      = grow_convex(0, 1.9e-3, 1.2)

fig, axes = plt.subplots(6, 5, figsize=(20, 18))
plt.subplots_adjust(hspace=0.4, wspace=0.4)
axes = axes.flatten()

orden_graficas = [
    (r'M1 M$\phi$s', 'M1'), (r'Infected M1 M$\phi$s', 'M1i'), (r'M2 M$\phi$s', 'M2'),
    (r'Infected M2 M$\phi$s', 'M2i'), ('HSCs', 'H'), ('Infected HSCs', 'Hi'),
    ('Th1 Cells', 'T1'), ('Th2 Cells', 'T2'), ('Fibroblast', 'f'), ('Myofibroblast', 'm'),
    ('ECM', 'rho'), (r'PDGF-$\beta$', 'G'), ('Hyaluronic Acid', 'HA'), (r'IL-1$\beta$', 'I1b'),
    ('IL-2', 'I2'), ('IL-4', 'I4'), ('IL-6', 'I6'), ('IL-10', 'I10'), ('IL-12', 'I12'),
    ('IL-13', 'I13'), (r'IFN-$\alpha$', 'Ia'), (r'INF-$\gamma$', 'Ig'), (r'TNF-$\alpha$', 'Ta'),
    (r'TGF-$\beta$', 'Tb'), ('MMP', 'Mp'), ('TIMP', 'Tp'), ('CCL3', 'C3'),
    ('Extracell HBV', 'Ve'), ('Intracell HBV', 'Vi_avg'), ('SCAR', 'S')
]

for i, (titulo, key) in enumerate(orden_graficas):
    ax = axes[i]
    ax.plot(t_graficas, data[key], linewidth=2.5, color='#1f77b4')

    ax.set_ylabel(titulo, fontsize=11, fontweight='bold')
    ax.set_xlabel("time (days)", fontsize=10)
    ax.set_xlim(0, 180)
    ax.set_xticks([0, 45, 90, 135, 180])

    ax.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
    ax.yaxis.offsetText.set_fontsize(10)

plt.tight_layout()
plt.show()
```
